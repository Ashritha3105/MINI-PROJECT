```python
import numpy as np
import matplotlib.pyplot as plt

def sharpe_ratio(returns, risk_free_rate):
 
    mean_return = np.mean(returns)
    std_dev = np.std(returns)
    sharpe_ratio = (mean_return - risk_free_rate) / std_dev

    return sharpe_ratio


returns = np.random.normal(loc=0.05, scale=0.1, size=1000)  

risk_free_rates = np.linspace(0, 0.1, 100) 


sharpe_ratios = [sharpe_ratio(returns, rfr) for rfr in risk_free_rates]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(risk_free_rates, sharpe_ratios)
plt.title('Sharpe Ratio vs. Risk-Free Rate')
plt.xlabel('Risk-Free Rate')
plt.ylabel('Sharpe Ratio')
plt.grid(True)
plt.show()
```

```python
#higher the sharpe ratio better is the risk-adjusted performance
#Sharpe ratio quantifies how much return an investment generates for the level of risk it takes on.
```
